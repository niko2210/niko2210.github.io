```python
M=0
m=0
def gcd(a,b):
    if a>b:
        M = a
        m = b
    elif b>a:
        M=b
        m=a
    elif b==a:
        return a
    if m == 0:
        return 'Error: input is zero'
        
    while M >m:
        M = M - m
        
    if m>M:
        return gcd(M,m)
    else:
        return m
```

```python
gcd(8,9)
```

```python
def is_prime(x):
    for y in range(1,x):
        if gcd(x,y) != 1:
            return False
    
    return True
```

```python
is_prime(18)
```

```python
for x in range(1,1000):
    if is_prime(x) == True:
        print(x)
```
